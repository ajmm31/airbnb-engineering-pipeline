# --- CAPA BRONZE: INGESTA CRUDA ---

Esta capa se encarga de:
- Ingestar datos en su formato original sin transformaciones
- Preservar la estructura y tipos de datos tal como vienen de la fuente
- Servir como punto de partida para trazabilidad y auditoría
- Almacenar datos en formato Delta para confiabilidad y performance

**Fuente de datos**: CSV con listados de propiedades de Airbnb
**Formato de salida**: Tabla Delta `bronze_airbnb`

```python
# 1. Definimos la ruta del archivo que acabas de subir
# Usando Unity Catalog Volume (workspace.default.data)
input_path = "/Volumes/workspace/default/data/airnb.csv"
output_table_name = "bronze_airbnb"
```

```python
# 2. Leemos el CSV
# Usamos inferSchema para que PySpark detecte automáticamente los tipos de datos iniciales
df_raw = (spark.read          
          .format("csv")          
          .option("header", "True")          
          .option("inferSchema", "True")          
          .load(input_path))
```

```python
# 3. Guardamos en formato Delta (Estándar de la arquitectura Medallion)
# 'overwrite' asegura que puedas correr esta celda varias veces mientras pruebas
(df_raw.write 
 .format("delta") 
 .mode("overwrite") 
 .saveAsTable(output_table_name))

print(f"¡Éxito! La tabla '{output_table_name}' ha sido creada en la capa Bronze.")
```

```python
# 4. Visualizamos una muestra de los datos
display(spark.table(output_table_name).limit(10))
```
